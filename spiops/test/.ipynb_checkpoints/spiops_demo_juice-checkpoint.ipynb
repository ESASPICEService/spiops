# spiops Tech Talk demo

This is a Jupyter Python Notebook to show a spiops demo. 
This could be your very own Python 3 environment, the first thing we will do is indicate that we want the Python package matplotlib to be output in the notebook and to import the spiops library

```python
%matplotlib inline
import spiops as spiops
```

Now, we will show the object oriented capabilities of sciops, aimed to easy the way that we interface with SPICE. 
We will load the JUICE SPICE Kernel Dataset, we will define a Time Window, Jupiter as a target and JUICE as an observer:

```python
spiops.load('/Users/mcosta/JUICE/kernels/mk/juice_crema_3_2_ops_local.tm')

interval = spiops.TimeWindow('2030-01-01T00:00:00', '2030-12-01T01:00:00', resolution=60*60)

jupiter = spiops.Target('JUPITER', time=interval, frame='IAU_JUPITER')
juice = spiops.Observer('JUICE', time=interval, target=jupiter)
```

spiops allows us to compute different geometric quantities for an observer for example:

- Distance to target
- Angle in between the Observer-Target and Observer Z Axis vectors

```python
juice.Plot('zaxis_target_angle', notebook=True)
juice.Plot('distance', notebook=True)
```

So, any geometrical quantity you need? The ones I need will be implemented, the ones you need can simply be requested.
But here comes the beauty of objects, what if I change the Time Window Object?

```python
interval.finish = '2030-05-05T12:00:00'
```

Let's see what happens:

```python
juice.Plot('zaxis_target_angle', notebook=True)
```

And this is even more beautiful:

```python
europa = spiops.Target('EUROPA', time=interval, frame='IAU_EUROPA')

juice.target = europa

juice.Plot('distance', notebook=True)
```

We even dare with some 3D:

```python
juice.Plot3D()
```
